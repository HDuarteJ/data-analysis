```python
import json
import pandas as pd
import requests
```

```python
# gets finacial data

def get_exchange_data(key, exchange='NYSE'):
    endpoint = f"https://eodhd.com/api/exchanges-list/?api_token={key}&fmt=json"
    call = requests.get(endpoint).text
    data = pd.DataFrame(json.loads(call))
    return data
```
